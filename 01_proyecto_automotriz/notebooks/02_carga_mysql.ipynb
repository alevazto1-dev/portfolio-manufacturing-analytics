# Carga del dataset automotriz a MySQL

Cargamos las 11 tablas del dataset automotriz desde CSVs a MySQL
usando pandas + SQLAlchemy. Una vez cargadas, hacemos análisis
con SQL puro desde DBeaver.

```python
import pandas as pd
from pathlib import Path
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

# Cargar variables de entorno
load_dotenv()

# Ruta al dataset
ruta_datos = Path("../data/raw/dataset_automotriz")

# Conexion a MySQL desde variables de entorno
engine = create_engine(
    f"mysql+pymysql://{os.getenv('MYSQL_USER')}:{os.getenv('MYSQL_PASSWORD')}@{os.getenv('MYSQL_HOST')}:{os.getenv('MYSQL_PORT')}/{os.getenv('MYSQL_DB')}"
)

# Verificar conexion
with engine.connect() as conn:
    print("Conexion exitosa a MySQL")
```

## Carga de tablas a MySQL

Cargamos las 11 tablas del dataset automotriz desde CSVs a MySQL.
Usamos `to_sql()` de pandas con `if_exists='replace'` para reemplazar
la tabla si ya existe.

```python
# Definir tablas a cargar
tablas = {
    "dim_turnos":            "dim_turnos.csv",
    "dim_proveedores":       "dim_proveedores.csv",
    "dim_materia_prima":     "dim_materia_prima.csv",
    "dim_maquinas":          "dim_maquinas.csv",
    "dim_productos":         "dim_productos.csv",
    "dim_operadores":        "dim_operadores.csv",
    "fact_produccion":       "fact_produccion.csv",
    "fact_paros":            "fact_paros.csv",
    "fact_defectos":         "fact_defectos.csv",
    "fact_mantenimiento":    "fact_mantenimiento.csv",
    "fact_inspeccion_calidad": "fact_inspeccion_calidad.csv"
}

# Cargar cada tabla
for nombre_tabla, archivo in tablas.items():
    df = pd.read_csv(ruta_datos / archivo)
    df.to_sql(nombre_tabla, engine, if_exists='replace', index=False)
    print(f"✓ {nombre_tabla}: {len(df):,} registros cargados")

print("\nCarga completa.")
```
